```python
from typing import Tuple

Vec3 = Tuple[float, float, float]

def diff(v0: Vec3, v1: Vec3) -> Vec3:
    return (v0[0] - v1[0], v0[1] - v1[1], v0[2] - v1[2])

def dot_product(v0: Vec3, v1: Vec3) -> float:
    return (v0[0] * v1[0]) + (v0[1] * v1[1]) + (v0[2] * v1[2])


def is_inside_triangle_barycentric(
    q: Vec3,
    p0: Vec3,
    p1: Vec3,
    p2: Vec3,
    eps: float = 1e-9,
) -> bool:
    """
    Return True if q is inside or on the boundary of triangle (p0, p1, p2).

    Suggested plan:
      1. Compute edge vectors from one vertex:
           e0 = p1 - p0
           e1 = p2 - p0
           r  = q  - p0
      2. Solve for barycentric coordinates.
      3. Check whether the coordinates lie inside the triangle.

    Notes:
      - You may want to treat points on edges/vertices as inside.
      - For degenerate triangles, returning False is usually reasonable.
      - `eps` is there for tolerance.
    """
    
    v0 = diff(p0, p2)
    v1 = diff(p1, p2)
    v2 = diff(q, p2)

    d00 = dot_product(v0, v0)
    d01 = dot_product(v0, v1)
    d02 = dot_product(v0, v2)
    d11 = dot_product(v1, v1)
    d12 = dot_product(v1, v2)

    det = (d11 * d00) - (d01 * d01)
    
    # Degenerate triangle
    if (abs(det) <  eps):
        return False
    
    a = ((d11 * d02) - (d01 * d12)) / det
    b = ((d12 * d00) - (d01 * d02)) / det

    return a >= 0 and b >= 0 and a + b <= 1
```
